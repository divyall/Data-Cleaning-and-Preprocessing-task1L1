```python
import pandas as pd
```

```python
df = pd.read_csv("2) Stock Prices Data Set (1) (1).csv")  # replace with your uploaded filename
```

```python
display(df.head())       # First 5 rows
display(df.info())       # Column types and non-null counts
display(df.describe())   # Statistics for numeric columns
```

```python
df.dropna(inplace=True)
```

```python
df['open'].fillna(df['open'].mean(), inplace=True)
df['high'].fillna(df['high'].mean(), inplace=True)
df['low'].fillna(df['low'].mean(), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df['symbol'] = df['symbol'].str.lower().str.strip()
```

```python
df.to_csv("cleaned_data.csv", index=False)

```

```python
print("Final Shape:", df.shape)
print("Data cleaning complete. Cleaned file saved.")
```
